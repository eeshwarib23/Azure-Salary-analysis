```
# Import necessary libraries
from pyspark.sql import SparkSession
from pyspark.sql.functions import col
import matplotlib.pyplot as plt
```

```
# Create a Spark session
spark = SparkSession.builder.appName("DataAnalysis").getOrCreate()


# 1. Age Distribution
data.select("Age").toPandas().hist(column="Age", bins=20)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```
# 2. Gender Distribution
gender_distribution = data.groupBy("Gender").count()
gender_distribution.show()
```

```
# 3. Education Level Analysis
education_level_distribution = data.groupBy("Education Level").count()
education_level_distribution.show()
```

```
# 4. Job Title Analysis
job_title_distribution = data.groupBy("Job Title").count().orderBy(col("count").desc())
job_title_distribution.show()
```

```
# 5. Years of Experience Analysis
avg_years_experience = data.agg({"Years of Experience": "avg"}).collect()[0][0]
median_years_experience = data.approxQuantile("Years of Experience", [0.5], 0.25)[0]
print(f"Average Years of Experience: {avg_years_experience}")
print(f"Median Years of Experience: {median_years_experience}")
```

```
# 6. Salary Analysis
salary_stats = data.describe("Salary").toPandas()
print(salary_stats)
```

```
# 7. Country and Race Distribution
country_distribution = data.groupBy("Country").count()
race_distribution = data.groupBy("Race").count()
country_distribution.show()
race_distribution.show()
```

```
# 8. Senior Position Analysis
senior_position_distribution = data.groupBy("Senior").count()
senior_position_distribution.show()
```
